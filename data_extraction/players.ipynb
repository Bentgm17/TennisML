```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings("ignore")
```

```python
def get_player_dataframe(df,w_l):
    stats_cols=[column for column in df.columns if w_l[0]+"_" in column]
    columns=['match_id',w_l+'_id','date']+stats_cols
    new_cols=['match_id','player_id','date']+[column[2:] for column in columns[3:]]
    new_df=df[columns]
    new_df.columns=new_cols
    new_df['label']=1 if w_l=='winner' else 0
    return new_df
```

```python
df=pd.read_csv("stats_df.csv",index_col=0).reset_index(drop=True)
winner_df=get_player_dataframe(df,'winner')
loser_df=get_player_dataframe(df,'loser')
players_df=pd.concat([winner_df,loser_df]).sort_values(by='player_id').reset_index(drop=True)
groupby_df=players_df.groupby('player_id')
display(groupby_df.first())
```
